```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# getting data
X, y = make_blobs(n_samples = 500, n_features=5, centers=3, cluster_std=1.5, random_state=42)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# applying pca and converted 5 columns into 2 columns
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
# to show in graph first convert it into df
df = pd.DataFrame(X_pca, columns=["PC1","PC2"])
df["label"] = y
sns.scatterplot(x=df["PC1"], y=df["PC2"],hue=df["label"], palette="rainbow")
```
